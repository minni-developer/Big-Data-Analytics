```python
import glob
import os
```

```python
# Specify the path to your CSV files
path = 'E:/SE-8/Project/EDUNONCORE-FEB2020'
```

```python

# Get a list of all CSV files in the directory
csv_files = glob.glob(os.path.join(path, "*.csv"))
```

```python
# Initialize an empty list to hold the DataFrames
dataframes = []
```

```python
# Function to process each chunk and filter relevant data
def process_chunk(chunk):
    # Ensure INDICATOR_ID is treated as string
    chunk['INDICATOR_ID'] = chunk['INDICATOR_ID'].astype(str)
    return chunk
```

```python

# Read each CSV file in chunks
for file in csv_files:
    try:
        # Process file in chunks
        chunk_list = []
        for chunk in pd.read_csv(file, dtype={'INDICATOR_ID': 'str'}, chunksize=100000):
            processed_chunk = process_chunk(chunk)
            chunk_list.append(processed_chunk)

        # Concatenate chunks after processing
        df = pd.concat(chunk_list, ignore_index=True)
        print(f"Columns in {file}: {df.columns.tolist()}")
        if 'INDICATOR_ID' in df.columns:
            dataframes.append(df)
        else:
            print(f"Warning: 'INDICATOR_ID' not found in {file}. Skipping this file.")
    except Exception as e:
        print(f"Error processing {file}: {e}")
        continue
```

```python

# Check if there are any DataFrames to merge
if dataframes:
    # Initialize the first DataFrame as the base for merging
    merged_df = dataframes[0]

    # Merge all other DataFrames onto the base DataFrame
    for df in dataframes[1:]:
        merged_df = pd.merge(merged_df, df, on='INDICATOR_ID', how='outer')  # Use 'inner' if you only want rows with matches in all files

    # Handle missing values (example: fill with 0)
    merged_df.fillna(0, inplace=True)

    # Save the merged DataFrame to a new CSV file
    merged_df.to_csv('merged_data.csv', index=False)
else:
    print("No DataFrames to merge.")
```
